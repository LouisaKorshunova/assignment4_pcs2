```python
#grph
def parse_fasta(file_content):
    sequences = {}
    header = None
    for line in file_content.strip().split("\n"):
        if line.startswith(">"):
            header = line[1:]
            sequences[header] = ""
        else:
            sequences[header] += line.strip()
    return sequences

def construct_overlap_graph(sequences, k):
    edges = []
    headers = list(sequences.keys())
    for i, src in enumerate(headers):
        suffix = sequences[src][-k:]
        for j, tgt in enumerate(headers):
            if i != j:
                prefix = sequences[tgt][:k]
                if suffix == prefix:
                    edges.append((src, tgt))
    return edges

def main():
    with open("/Users/makbuk/Downloads/rosalind_grph.txt", "r") as file:
        file_content = file.read()
    sequences = parse_fasta(file_content)
    k = 3
    edges = construct_overlap_graph(sequences, k)
    for edge in edges:
        print(f"{edge[0]} {edge[1]}")

if __name__ == "__main__":
    main()
```

```python
#long
def read_fasta():
    with open("/Users/makbuk/Downloads/rosalind_long.txt", "r") as file:
        sequences = []
        current_sequence = []
        for line in file:
            if line.startswith('>'):
                if current_sequence:
                    sequences.append(''.join(current_sequence))
                    current_sequence = []
            else:
                current_sequence.append(line.strip())
        if current_sequence:
            sequences.append(''.join(current_sequence))
    return sequences


def find_overlap(a, b):
    max_overlap = 0
    overlap_index = 0
    for i in range(1, len(a)):
        if b.startswith(a[i:]):
            overlap_index = i
            max_overlap = len(a) - i
            break
    return max_overlap, overlap_index


def shortest_superstring(strings):
    while len(strings) > 1:
        max_overlap = -1
        pair_to_merge = (0, 0)
        merge_string = ""
        for i in range(len(strings)):
            for j in range(len(strings)):
                if i != j:
                    overlap, index = find_overlap(strings[i], strings[j])
                    if overlap > max_overlap:
                        max_overlap = overlap
                        pair_to_merge = (i, j)
                        merge_string = strings[i] + strings[j][overlap:]
        i, j = pair_to_merge
        strings.pop(max(i, j))
        strings.pop(min(i, j))
        strings.append(merge_string)
    return strings[0]


def solve_superstring_problem():
    dna_strings = read_fasta()
    result = shortest_superstring(dna_strings)
    print(result)

solve_superstring_problem()
```

```python
#deg
def calculate_degrees(n, edges):
    degrees = [0] * n
    for u, v in edges:
        degrees[u - 1] += 1
        degrees[v - 1] += 1
    return degrees

def main():
    with open("/Users/makbuk/Downloads/rosalind_deg.txt", "r") as file:
        data = file.readlines()
    n, m = map(int, data[0].strip().split())
    edges = [tuple(map(int, line.strip().split())) for line in data[1:]]
    degree_array = calculate_degrees(n, edges)
    print(" ".join(map(str, degree_array)))

if __name__ == "__main__":
    main()
```

```python
#cc
def solve_connected_components():
    with open("/Users/makbuk/Downloads/rosalind_cc.txt", 'r') as f:
        n, m = map(int, f.readline().strip().split())
        adj_list = {i: [] for i in range(1, n+1)}
        for _ in range(m):
            u, v = map(int, f.readline().strip().split())
            adj_list[u].append(v)
            adj_list[v].append(u)
    visited = [False] * (n + 1)

    def dfs(v):
        stack = [v]
        while stack:
            node = stack.pop()
            if not visited[node]:
                visited[node] = True
                for neighbor in adj_list[node]:
                    if not visited[neighbor]:
                        stack.append(neighbor)
    components_count = 0
    for i in range(1, n + 1):
        if not visited[i]:
            dfs(i)
            components_count += 1
    print(components_count)

solve_connected_components()
```

```python
#bfs
from collections import deque

def solve_single_source_shortest_distances():
    with open("/Users/makbuk/Downloads/rosalind_bfs.txt", 'r') as f:
        n, m = map(int, f.readline().strip().split())
        adj_list = {i: [] for i in range(1, n+1)}
        for _ in range(m):
            u, v = map(int, f.readline().strip().split())
            adj_list[u].append(v)
    D = [-1] * (n + 1)
    D[1] = 0
    queue = deque([1])
    
    while queue:
        current = queue.popleft()
        for neighbor in adj_list[current]:
            if D[neighbor] == -1:
                D[neighbor] = D[current] + 1
                queue.append(neighbor)
    print(" ".join(map(str, D[1:])))

solve_single_source_shortest_distances()
```

```python
#dij
import heapq
import sys

def dijkstra(graph, n):
    D = [sys.maxsize] * (n + 1)
    D[1] = 0
    pq = [(0, 1)]

    while pq:
        current_dist, u = heapq.heappop(pq)
        if current_dist > D[u]:
            continue
        for v, weight in graph[u]:
            distance = current_dist + weight
            if distance < D[v]:
                D[v] = distance
                heapq.heappush(pq, (distance, v))
    for i in range(1, n + 1):
        if D[i] == sys.maxsize:
            D[i] = -1

    return D[1:]

def read_input(file_name):
    with open(file_name, 'r') as file:
        n, m = map(int, file.readline().strip().split())
        graph = {i: [] for i in range(1, n + 1)}
        for _ in range(m):
            u, v, w = map(int, file.readline().strip().split())
            graph[u].append((v, w))

    return graph, n

def main():
    graph, n = read_input('/Users/makbuk/Downloads/rosalind_dij.txt')
    result = dijkstra(graph, n)
    print(' '.join(map(str, result)))

if __name__ == "__main__":
    main()
```

```python
#dag
def solve_acyclic():
    with open("/Users/makbuk/Downloads/rosalind_dag.txt", 'r') as f:
        k = int(f.readline().strip())
        results = []
    
        for _ in range(k):
            line = f.readline().strip()
            while not line:
                line = f.readline().strip()
            n, m = map(int, line.split())
            adj_list = {i: [] for i in range(1, n + 1)}
        
            for _ in range(m):
                edge_line = f.readline().strip()
                while not edge_line:
                    edge_line = f.readline().strip()
                
                u, v = map(int, edge_line.split())
                adj_list[u].append(v)
            
            def dfs(v, visited, rec_stack):
                visited[v] = 1
                rec_stack[v] = 1
                
                for neighbor in adj_list[v]:
                    if visited[neighbor] == 0:
                        if dfs(neighbor, visited, rec_stack):
                            return True
                    elif rec_stack[neighbor] == 1:
                        return True
                
                rec_stack[v] = 0
                return False
            
            visited = [0] * (n + 1)
            rec_stack = [0] * (n + 1)
            cycle_found = False

            for vertex in range(1, n + 1):
                if visited[vertex] == 0:
                    if dfs(vertex, visited, rec_stack):
                        cycle_found = True
                        break
            
            results.append("1" if not cycle_found else "-1")
        
        print(" ".join(results))

solve_acyclic()
```

```python
#bf
def solve_bellman_ford():
    with open("/Users/makbuk/Downloads/rosalind_bf.txt", 'r') as f:
        n, m = map(int, f.readline().strip().split())
        edges = []
        
        for _ in range(m):
            u, v, weight = map(int, f.readline().strip().split())
            edges.append((u, v, weight))
        
        D = [float('inf')] * (n + 1)
        D[1] = 0
        
        for _ in range(n - 1):
            for u, v, weight in edges:
                if D[u] != float('inf') and D[u] + weight < D[v]:
                    D[v] = D[u] + weight
        
        result = []
        for i in range(1, n + 1):
            if D[i] == float('inf'):
                result.append("x")
            else:
                result.append(str(D[i]))
        
        print(" ".join(result))

solve_bellman_ford()
```

```python
#cte
import heapq

def dijkstra(graph, n, start):
    dist = [float('inf')] * (n + 1)
    dist[start] = 0
    pq = [(0, start)]
    
    while pq:
        current_dist, u = heapq.heappop(pq)
        
        if current_dist > dist[u]:
            continue
        
        for v, weight in graph[u]:
            new_dist = current_dist + weight
            if new_dist < dist[v]:
                dist[v] = new_dist
                heapq.heappush(pq, (new_dist, v))
    
    return dist

def solve_shortest_cycle():
    with open("/Users/makbuk/Downloads/rosalind_cte.txt", 'r') as f:
        k = int(f.readline().strip())
        results = []
        
        for _ in range(k):
            n, m = map(int, f.readline().strip().split())
            edges = []
            graph = {i: [] for i in range(1, n + 1)}
            
            for _ in range(m):
                u, v, w = map(int, f.readline().strip().split())
                edges.append((u, v, w))
                graph[u].append((v, w))
            
            u, v, w = edges[0]
            graph[u] = [edge for edge in graph[u] if edge[0] != v]
            dist_from_v = dijkstra(graph, n, v)

            if dist_from_v[u] == float('inf'):
                results.append("-1")
            else:
                cycle_length = dist_from_v[u] + w
                results.append(str(cycle_length))
        
        print(" ".join(results))

solve_shortest_cycle()
```

```python
#sdag
def solve_dag_shortest_path():
    with open("/Users/makbuk/Downloads/rosalind_sdag.txt", 'r') as f:
        n, m = map(int, f.readline().strip().split())
        adj_list = {i: [] for i in range(1, n + 1)}
        
        for _ in range(m):
            u, v, weight = map(int, f.readline().strip().split())
            adj_list[u].append((v, weight))
        
        topological_order = []
        visited = [False] * (n + 1)

        def dfs(u):
            visited[u] = True
            for v, _ in adj_list[u]:
                if not visited[v]:
                    dfs(v)
            topological_order.append(u)
        
        for i in range(1, n + 1):
            if not visited[i]:
                dfs(i)
        
        topological_order.reverse()

        D = [float('inf')] * (n + 1)
        D[1] = 0
        
        for u in topological_order:
            if D[u] != float('inf'):
                for v, weight in adj_list[u]:
                    if D[u] + weight < D[v]:
                        D[v] = D[u] + weight

        result = []
        for i in range(1, n + 1):
            if D[i] == float('inf'):
                result.append("x")
            else:
                result.append(str(D[i]))
        
        print(" ".join(result))

solve_dag_shortest_path()
```
